```python
msno.bar(planets);
```



*  Yukarıdaki grafik değişkenlerdeki eksikliği oransal olarak ifade etmektedir.
*  Grafiğin üst tarafındaki değerler eksik olmayan veri değerleridir.



Eksik verilerin yerine aritmetik ortalama ,mod, medyan ya da istenilen herhangi değerler atanabilir.

```python
planets["mass"].fillna(0,inplace=True)
planets
#mass sütunundaki eksik verilerin yerine 0 girilmesi.
```

```python
planets["distance"].fillna(planets["distance"].mean(),inplace=True)
planets
```

Distance değişkenindeki boş yerlere ortalama değer atandı.
